```python
 import pandas as pd
import numpy as np
from numpy import nan as NA
```

## Filtering out missing values

method 1-

```python
data= pd.Series([1,NA,3.5,NA,7])
data
```

```python
data.dropna()     # dropping missing values from the table
```

```python
data[data.notnull()]   # same as before
```

method 2-

```python
data= pd.DataFrame([[1.0, 6.5, 3.0], [1.0, NA, NA], [NA,NA,NA], [NA, 6.5, 3.0]])
data
```

```python
cleaned= data.dropna()    # drop all rows which have NA in them
cleaned
```

```python
cleaned= data.dropna(how='all')          #  drop rows which have all elements as NA
cleaned
```

```python
data[4]= NA       # add 4th column with all values as NA
data
```

```python
cleaned= data.dropna(axis=1, how='all')    # axis 1-row, axis 0-column
cleaned
```

method 3-

```python
df= pd.DataFrame(np.random.randn(7,3))
df.iloc[:4,1]= NA            # iloc[a,b] -  a- row, b- column
df.iloc[:2,2]= NA
df
```

```python
df.dropna()                     # drop all rows which has NA
```

```python
df.dropna(thresh=2)              # row must have atleast threshold non-NA values
```

## Filling in Missing Data


method 1-

```python
df
```

```python
df.fillna(0)                    # fillna(a) replaces NA with constant a in dataframe
```

method 2-

```python
df.fillna({1: 0.5, 2: 0})        #  column 1- replace with 0.5    column 2- replace with 0
```

```python
df.fillna(0, inplace=True)      # modifies in the origional dataframe
df
```

```python
df= pd.DataFrame(np.random.randn(6,3))
df.iloc[2:,1]= NA
df.iloc[4:,2]= NA
df
```

```python
df.fillna(method='ffill')      # fill the NA columns with value just above the NA cell in the column
```

```python
df.fillna(method='ffill', limit=2)    # fill max 2 columns
```

##Transforming data usng a Function or Mapping

```python
data= pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                             'Pastrami', 'corned beef', 'Bacon',
                             'pastrami', 'honey ham', 'nova lox'],
                    'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal= {
                  'bacon': 'pig',
                  'pulled pork': 'pig',
                  'pastrami': 'cow',
                  'corned beef': 'cow',
                  'honey ham': 'pig',
                  'nova lox': 'salmon'
                }
```

```python
lowercased= data['food'].str.lower()
lowercased
```

mapping

```python
data['animal']= lowercased.map(meat_to_animal)
data
```

```python
data['food'].map(lambda x:meat_to_animal[x.lower()])
```

##Replacing Values

```python
data= pd.Series([1.,-999,2.,-999,-1000.,3.])
data
```

```python
data.replace(-999,np.nan)   # replace a value by a value
```

```python
data.replace([-999,-1000], np.nan)    # replace a set of values by a value
```

```python
data.replace([-999,-1000],[np.nan,0])    # replace a set of values by a set of values
```

```python
data.replace({-999: np.nan, -1000: 0})    # using dictionary
```

##Renaming Axis Indexes

```python
data= pd.DataFrame(np.arange(12).reshape((3,4)),
                    index=['Ohio', 'Colorado', 'New York'],
                    columns=['one','two','three','four'])
data
```

```python
transform= lambda x: x[:4].upper()
data.index.map(transform)
```

```python
data.index= data.index.map(transform)
data
```

```python
data.rename(index=str.title, columns=str.upper)
```

```python
data.rename(index={'OHIO': 'INDIANA'},
            columns={'three':'peekaboo'})
```

```python
data.rename(index={'OHIO':'INDIANA'}, inplace=True)
data
```

##Discretization and Binning

```python
ages= [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

```python
bins= [18, 25, 35, 60, 100]
```

```python
cats= pd.cut(ages,bins)
cats
```

```python
cats.codes
```

```python
cats.categories
```

```python
pd.value_counts(cats)
```

```python
pd.cut(ages,[18,26,36,61,100], right=False)
```

```python
group_names= ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
```

```python
pd.cut(ages, bins, labels=group_names)
```

```python
data= np.random.rand(20)
```

```python
pd.cut(data, 4, precision=2)
```

```python
data= np.random.randn(1000) # Normally distributed
```

```python
cats= pd.qcut(data, 4)  # cut into quartiles
cats
```
